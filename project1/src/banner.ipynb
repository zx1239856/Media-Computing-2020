```python
from PIL import Image
from graphcut import GraphCut
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt
import numpy as np
```

```python
names = 'cvpr'
images = []
for n in names:
    img = Image.open(f'../output/banner/{n}.png').convert('RGB')
    images.append(img)
    plt.figure()
    plt.axis('off')
    plt.imshow(img)
```

```python
out_h, out_w = images[0].height, images[0].width * 3 - 50
blender = GraphCut(np.array(images[0]), (out_h, out_w))
blender.insert_patch(0, 0, images[0].width, images[1].height)
blender.set_patch(np.array(images[1]))
blender.insert_patch(350, 0, images[1].width, images[1].height)
blender.set_patch(np.array(images[2]))
blender.insert_patch(700, 0, images[2].width, images[2].height)
blender.set_patch(np.array(images[3]))
blender.insert_patch(1000, 0, images[3].width, images[3].height)
plt.figure()
plt.axis('off')
plt.imshow(blender.output)
Image.fromarray(blender.output).save('../output/banner/cvpr.png')
```
